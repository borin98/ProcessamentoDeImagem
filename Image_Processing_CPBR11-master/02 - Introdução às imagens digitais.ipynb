# Introdução às imagens digitais

Uma imagem digital pode ser representada por uma matriz bidimensional, onde os seus elementos são chamados de pixels (picture elements).

Iremos utilizar a matriz disponível no ndarray NumPy para representar as imagens. A vantagem é que todas as operações disponíveis para processamento matricial podem ser utilizados como processamento de imagens.

Vamos trabalhar basicamente em 3 tipos representações das imagens:

** Binárias ** = cada voxel pode ter o valor True ou False

** uint8 ** = (mais utilizada) cada voxel pode assumir números inteiros entre 0 - 255

** float ** = cada voxel contém um valor entre 0.0 e 1.0

## Leitura de Imagens

```python
# https://matplotlib.org/2.0.2/users/image_tutorial.html

import matplotlib.pyplot as plt
import matplotlib.image as mpimg

img=mpimg.imread('figures/cameraman.tif')
```

### Exibindo a imagem como array

```python
print("Número De Linhas = ", img.shape[0])
print("Número De Colunas = ", img.shape[1])
print("Dimensão = ", img.ndim)
print("Formato = ", img.shape)
print("datatype = ", img.dtype )
print("img.max()", img.max())
print("img.min()", img.min())

print(img)
```

## Plotando a imagem

```python
imgplot = plt.imshow(img, 'gray')
plt.show()
```

## Convertendo a imagem para float

```python
img_float = img / 255.
```

```python
print("Número De Linhas = ", img_float.shape[0])
print("Número De Colunasa = ", img_float.shape[1])
print("Dimensão = ", img_float.ndim)
print("Formato = ", img_float.shape)
print("datatype = ", img_float.dtype )
print("img.max()", img_float.max())
print("img.min()", img_float.min())

print(img_float)
```

```python
imgplot = plt.imshow(img_float, 'gray')
plt.show()
```

## Binarizando a imagem

```python
threshold = 60
img_BW = img>threshold
```

```python
print("Número De Linhas = ", img_BW.shape[0])
print("Número De Colunas = ", img_BW.shape[1])
print("Dimensão = ", img_BW.ndim)
print("Formato = ", img_BW.shape)
print("datatype = ", img_BW.dtype )
print("img.max()", img_BW.max())
print("img.min()", img_BW.min())

print(img_BW)
```

```python
imgplot = plt.imshow(img_BW, 'gray')
plt.show()
```

# Imagens coloridas

A difereça entre as imagens em tons de cinza e das imagens coloridas é que esta útima possui canais de cores.
O padrão mais utilizado é o RGB (Red, Green, Blue), que é exatamente o padrão usado nos monitores.

Com a adição do canal de cores, o array que representa a imagem agora passa a ter uma terceira dimensão.

```python
img=mpimg.imread('figures/Lenna.png')
```

```python
print("Número De Linhas = ", img.shape[0])
print("Número De Colunas = ", img.shape[1])
print("Dimensão = ", img.ndim)
print("Formato = ", img.shape)
print("datatype = ", img.dtype )
print("img.max()", img.max())
print("img.min()", img.min())

print(img)
```

```python
imgplot = plt.imshow(img)
plt.show()
```

## Exibindo cada canal separadamente

```python
img_R = img[:,:,0]
img_G = img[:,:,1]
img_B = img[:,:,2]
```

```python
plt.figure(figsize=(24, 12))

plt.subplot(1, 3, 1)
plt.imshow(img_R, 'gray')

plt.subplot(1, 3, 2)
plt.imshow(img_G, 'gray')

plt.subplot(1, 3, 3)
plt.imshow(img_B, 'gray')

plt.show()
```

## Convertendo a imagem para tons de cinza

```python
img_gray = (img_R + img_G + img_B) / 3
```

```python
print("Número De Linhas = ", img_gray.shape[0])
print("Número De Colunas = ", img_gray.shape[1])
print("Dimensão = ", img_gray.ndim)
print("Formato = ", img_gray.shape)
print("datatype = ", img_gray.dtype )
print("img.max()", img_gray.max())
print("img.min()", img_gray.min())

print(img_gray)
```

```python
imgplot = plt.imshow(img_gray, 'gray')
plt.show()
```

# Modificações simples nas imagens

### Quadriculado sobreposto na imagem

```python
img=mpimg.imread('figures/CP.jpg')
img_gray = img[:,:,1]
```

```python
print("Número De Linhas = ", img_gray.shape[0])
print("Número De Colunas = ", img_gray.shape[1])
print("Dimensão = ", img_gray.ndim)
print("Formato = ", img_gray.shape)
print("datatype = ", img_gray.dtype )
print("img.max()", img_gray.max())
print("img.min()", img_gray.min())

imgplot = plt.imshow(img_gray, 'gray')
plt.show()
```

```python
import numpy as np

delta = 20

g = np.copy(img_gray)
g[::delta,:] = 255
g[:,::delta] = 255
```

```python
imgplot = plt.imshow(g, 'gray')
plt.show()
```

## slicing

```python
img=mpimg.imread('figures/CP.jpg')
img_gray = img[:,:,1]
```

```python
slice_y = 3
slice_x = 5

g = np.zeros(img_gray.shape)
g[::slice_x,:] = 1
g[:,::slice_y] = 1

g = g * img_gray

imgplot = plt.imshow(g, 'gray')
plt.show()
```

```python
# Parcela descartada
imgplot = plt.imshow(img_gray - g, 'gray')
plt.show()
```

```python
img_sliced = img_gray[::slice_x,::slice_y]

imgplot = plt.imshow(img_sliced, 'gray')
plt.show()
```
